Correlation Matrix Heatmap

```python
# Computing the Correlation Matrix 
spike_cols = [col for col in music_data.columns if 'mean' in col] 

# Set up the matplotlib figure 
f, ax = plt.subplots(figsize=(16, 11)) 

# Draw the heatmap with the mask and correct aspect ratio 
sns.heatmap(music_data[spike_cols].corr(), cmap='YlGn') 

plt.title('Heatmap for MEAN variables', fontsize = 20) 
plt.xticks(fontsize = 10) 
plt.yticks(fontsize = 10)
```

Label Encoding and Feature Scaling

```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 
music_data['label'] = label_encoder.fit_transform(music_data['label'])
```

```python
X = music_data.drop(['label','filename'],axis=1) 
y = music_data['label']
```

```python
cols = X.columns 
minmax = preprocessing.MinMaxScaler() 
np_scaled = minmax.fit_transform(X) 

# new data frame with the new scaled data. 
X = pd.DataFrame(np_scaled, columns = cols)
```

Splitting Data

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=111) 
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

Classifier Models

```python
from sklearn.metrics import accuracy_score 
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier 
from sklearn.linear_model import LogisticRegression 
import catboost as cb 
from xgboost import XGBClassifier 

knn = KNeighborsClassifier(n_neighbors=5)
dtc = DecisionTreeClassifier(random_state=7)
rf = RandomForestClassifier(n_estimators=1000, max_depth=10, random_state=0) 
lgc = LogisticRegression(random_state=0)
cbc = cb.CatBoostClassifier(verbose=0, eval_metric='Accuracy', loss_function='MultiClass') 
xgb = XGBClassifier(n_estimators=1000, learning_rate=0.05) 

for clf in (knn, dtc, rf, lgc, cbc, xgb): 
	clf.fit(X_train, y_train) 
	preds = clf.predict(X_test) 
	print(clf.__class__.__name__,accuracy_score(y_test, preds)*100)
```

Neural Network Model

```python
import tensorflow.keras as keras  # type: ignore
from tensorflow.keras import Sequential  # type: ignore
from tensorflow.keras.layers import *  # type: ignore

model = Sequential() 

model.add(Flatten(input_shape=(58,))) # type: ignore
model.add(Dense(256, activation='relu')) # type: ignore
model.add(BatchNormalization()) # type: ignore
model.add(Dense(128, activation='relu')) # type: ignore
model.add(Dropout(0.3)) # type: ignore
model.add(Dense(10, activation='softmax')) # type: ignore
model.summary()
```

```python
# compile the model 
adam = keras.optimizers.Adam(lr=1e-4) 
model.compile(optimizer=adam, 
			loss="sparse_categorical_crossentropy", 
			metrics=["accuracy"]) 

hist = model.fit(X_train, y_train, 
				validation_data = (X_test,y_test), 
				epochs = 100, 
				batch_size = 32)
```

```python
test_error, test_accuracy = model.evaluate(X_test, y_test, verbose=1) 
print(f"Neural Network Test accuracy: {test_accuracy*100}")
```

```python
fig, axs = plt.subplots(2,figsize=(10,10)) 

# accuracy 
axs[0].plot(hist.history["accuracy"], label="train") 
axs[0].plot(hist.history["val_accuracy"], label="test")	 
axs[0].set_ylabel("Accuracy") 
axs[0].legend() 
axs[0].set_title("Accuracy") 
	
# Error 
axs[1].plot(hist.history["loss"], label="train") 
axs[1].plot(hist.history["val_loss"], label="test")	 
axs[1].set_ylabel("Error") 
axs[1].legend() 
axs[1].set_title("Error") 
	
plt.show()
```
